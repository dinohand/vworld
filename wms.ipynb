```python
import folium

vworld_key='1AA256E1-B218-38A1-9BFF-A0032E752166'
wtms_url = f"https://api.vworld.kr/req/wmts/1.0.0/{vworld_key}/Base/{{z}}/{{y}}/{{x}}.png"
wts_url = f"https://api.vworld.kr/req/wms?"


layer_map_name = "JK's Map"


# Map을 생성한다
# folium 기본 web map tile로 표시된다다
m = folium.Map(
    location = [ 37.575560576, 127.058889404 ],
    zoom_start = 18
)

folium.TileLayer(
    tiles=wtms_url,
    attr= "JK's Tile",
    name=layer_map_name
).add_to(m)


# m.save('test.html')



## Tile을 v-world 타일로 변경한다 
## layers의 합이 4개를 초과할 수 없다다
folium.WmsTileLayer(
    url=wts_url,
    layers='lt_c_landinfobasemap',     ## "LX맵(편집지적도"는 zoom:18 에서 보임
    request='GetMap',
    version='1.3.0',
    height = 256,
    width=256,
    fmt='image/png',
    key=vworld_key,
    transparent=True,
    name='LX맵(편집지적도)-zoom:18'
    
).add_to(m)

## 교통
folium.WmsTileLayer(
    url=wts_url,
    layers='lt_p_utiscctv,lt_p_moctnode,lt_l_moctlink', ### ,lt_l_n3a0020000',   
    request='GetMap',
    version='1.3.0',
    height = 256,
    width=256,
    fmt='image/png',
    key=vworld_key,
    transparent=True,
    name='교통'
).add_to(m)

# ## 문화/관광
# folium.WmsTileLayer(
#     url=wts_url,
#     layers='lt_p_dgtouristinfo,lt_p_tradsijang,lt_c_uo301, lt_p_dgmuseumart, lt_c_uo501, lt_p_smalllibrary',  
#     request='GetMap',
#     version='1.3.0',
#     height = 256,
#     width=256,
#     fmt='image/png',
#     key=vworld_key,
#     transparent=True,
#     name='문화/관광'
# ).add_to(m)


folium.LayerControl().add_to(m)


m.save('test.html')  ## 라이브 서버에서 호출하면 됨
```
